```python
# read the Modern India book file

india = open('Modern India.txt')
text = india.read()
```

```python
#import Freq Dist library

from nltk import FreqDist
```

```python
# printing the length of the text
# before tokenization

print(len(text))
```

```python
#perform tokenization

indiaTokens = nltk.word_tokenize(text)
```

```python
indiawords = [w.lower() for w in indiaTokens]
```

```python
#print length of tokenized text

print(len(indiawords))
```

```python
# top 50 freq words

ndist = FreqDist(indiawords)
nitems = ndist.most_common(50)
for item in nitems: 
    print(item[0],'\t',item[1])
```

```python
# using regular expression we remove the non alphabetical characters

import re 
pattern = re.compile('^[^a-z]+$')
```

```python
# function to return the list of words which 
# do not include non alphabetical characters

def alpha_filter(w):
    pattern = re.compile('^[^a-z]+$')
    if(pattern.match(w)):
        return True
    else:
        return False    
```

```python
indiaalphawords = [w for w in indiawords if not alpha_filter(w)]
print(len(indiaalphawords))
print(indiaalphawords)
```

```python
# top 50 most frequent words
# after removing non alphabetical characters

ndist = FreqDist(indiaalphawords)
nitems = ndist.most_common(50)
for item in nitems:
    print(item[0], '\t',item[1])
```

```python
# remove most frequent stop words from our list

stopwords = nltk.corpus.stopwords.words('english')
print('number of stopwords:', len(stopwords))
print(stopwords)
```

```python

stoppedindiawords = [w for w in indiaalphawords if not w in stopwords]
```

```python
#print updated freq list
indiadist = FreqDist(stoppedindiawords)
indiaitems = indiadist.most_common(50)
for item in indiaitems:
    print(item)
```

```python
# create an empty list 
wordsFiltered = []
# add the stopwords to this existing list
for w in stopwords:
        wordsFiltered.append(w)
#add custom stopwords
wordsFiltered.append('much')
wordsFiltered.append('made')
wordsFiltered.append('may')
wordsFiltered.append('many')
wordsFiltered.append('even')
print(wordsFiltered)
```

```python
stoppedindiawords = [w for w in indiaalphawords if not w in wordsFiltered]
# Print the updated list of frequencies without stop words
indiadist = FreqDist(stoppedindiawords)
indiaitems = indiadist.most_common(50)
for item in indiaitems:
    print(item)
```
