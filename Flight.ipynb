```python
import nltk
import os
```

```python
os.getcwd()
```

```python
#read the flight book file

flight = open('Flight.txt')
text = flight.read()
```

```python
# import the FreqDist library

from nltk import FreqDist
```

```python
# printing the length of the text
# before tokenization

print(len(text))
```

```python
# performing tokenization on the Flight text

flightTokens = nltk.word_tokenize(text)
```

```python
# convert the tokenized words to lower case

flightwords = [w.lower() for w in flightTokens]
```

```python
# print the length of tokenized words

print(len(flightwords))
```

```python
# Print after tokenization

print(flightwords)
```

```python
# top 50 most frequent words
# without any filters

ndist = FreqDist(flightwords)
nitems = ndist.most_common(50)
for item in nitems:
    print(item[0], '\t',item[1])
```

```python
# using regular expression we remove the non alphabetical characters

import re 
pattern = re.compile('^[^a-z]+$')
```

```python
# function to return the list of words which 
# do not include non alphabetical characters

def alpha_filter(w):
    pattern = re.compile('^[^a-z]+$')
    if(pattern.match(w)):
        return True
    else:
        return False    
```

```python
flightalphawords = [w for w in flightwords if not alpha_filter(w)]
print(len(flightalphawords))
print(flightalphawords)
```

```python
# top 50 most frequent words
# after removing non alphabetical characters

ndist = FreqDist(flightalphawords)
nitems = ndist.most_common(50)
for item in nitems:
    print(item[0], '\t',item[1])
```

```python
# remove most frequent stop words from our list

stopwords = nltk.corpus.stopwords.words('english')
print('number of stopwords:', len(stopwords))
print(stopwords)
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stoppedflightwords = [w for w in flightalphawords if not w in stopwords]
```

```python
# Print the updated list of frequencies without stop words
flightdist = FreqDist(stoppedflightwords)
flightitems = flightdist.most_common(50)
for item in flightitems:
    print(item)
```

```python
# create an empty list 
wordsFiltered = []
# add the stopwords to this existing list
for w in stopwords:
        wordsFiltered.append(w)
#add custom stopwords
wordsFiltered.append('much')
wordsFiltered.append('made')
print(wordsFiltered)
```

```python
stoppedflightwords = [w for w in flightalphawords if not w in wordsFiltered]
# Print the updated list of frequencies without stop words
flightdist = FreqDist(stoppedflightwords)
flightitems = flightdist.most_common(50)
for item in flightitems:
    print(item)
```

```python
#Bigrams

flightbigrams = list(nltk.bigrams(flightwords))
print(flightbigrams)
```

```python
from nltk.collocations import *
```

```python
bigram_measure = nltk.collocations.BigramAssocMeasures()
```

```python
finder = BigramCollocationFinder.from_words(flightwords)
scored = finder.score_ngrams(bigram_measure.raw_freq)
```

```python
for bscore in scored[:50]:
    print(bscore)
```

```python
finder.apply_word_filter(alpha_filter)
scored = finder.score_ngrams(bigram_measure.raw_freq)
for bscore in scored[:50]:
    print(bscore)
```

```python
finder.apply_word_filter(lambda w: w in stopwords)
scored = finder.score_ngrams(bigram_measure.raw_freq)
for bscore in scored[:50]:
    print(bscore)
```

```python
finder2 = BigramCollocationFinder.from_words(flightwords)
finder2.apply_freq_filter(5)
scored = finder2.score_ngrams(bigram_measure.raw_freq)
for bscore in scored[:50]:
    print(bscore)
```

```python
finder.apply_ngram_filter(lambda w1,w2: len(w1)<2)
scored = finder.score_ngrams(bigram_measure.raw_freq)
for bscore in scored[:50]:
    print(bscore)
```
